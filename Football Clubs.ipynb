# Real Madrid DataFrame

```python

df2=pd.read_csv('E:\Pyhton codes\data\data.csv')
def clean_team_dataframe(df2,str=None):
    df2=df2.where(df2['Club']==str).dropna(how='all')
    df2.set_index('Name',inplace=True)
    df2.drop(['Unnamed: 0','ID','Photo','Flag','Club Logo'],axis=1,inplace=True)
    return df2

```

```python
Real_Madrid=pd.read_csv('E:\Pyhton codes\data\data.csv')
clean_team_dataframe(Real_Madrid,str='Real Madrid')
```

```python
def get_mean_age(Real_Madrid,str='Real Madrid'):
    
    df=clean_team_dataframe(Real_Madrid,str)
    mean_age=df['Age'].mean()
    return mean_age

get_mean_age(Real_Madrid,str='Real Madrid')
```

```python
Barcelona=pd.read_csv('E:\Pyhton codes\data\data.csv')
clean_team_dataframe(Barcelona,str='FC Barcelona')
```

```python
get_mean_age(Barcelona,str='FC Barcelona')
```

```python

def get_mean_overall(Real_Madrid,str='Real Madrid'):
    
    df=clean_team_dataframe(Real_Madrid,str)
    mean_overall=df['Overall'].mean()
    return mean_overall

get_mean_overall(Real_Madrid,str='Real Madrid')
```

```python
get_mean_overall(Barcelona,str='FC Barcelona')
```
